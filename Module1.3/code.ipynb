Basic operations

```python
import numpy as np

arr=np.array([2,3,4,5,6,7])

print(arr+3)
print(arr/2)
```

array versus array operation

```python
a=np.array([1,2,3])
b=np.array([2,3,4])

print(a+b)
print(a*b)
```

methods

```python
print(arr.sum())
print(arr.max())
print(arr.min())
print(arr.mean())
```

standard deviation and variance

```python
print(arr.std())
print(arr.var())
```

square and square root

```python
print(np.square(arr))
print(np.sqrt(arr))
```

practice part

```python
#basic operation
import numpy as np
arr=np.array([[1,2,3,4,5,6,7],[2,3,4,5,6,7,8]])

print(arr+1)
print(arr*2)
print(arr/2)

#function
print(arr.sum())
print(arr.max())
print(arr.mean())
print(arr.std())
print(arr.var())
print(arr.sum(axis=0))
print(arr.sum(axis=1))
print(np.sqrt(arr))
print(np.square(arr))
```
