array filtering

```python
import numpy as np
arr=np.array([10,9,8,49,20,45,1,2,43])
print(arr>2)
print(arr[arr>3])
```

multiple condition

```python
import numpy as np
arr=np.array([10,9,8,49,20,45,1,2,43])
print(arr[(arr>9) & (arr<30)])
print(arr[(arr<5) | (arr>30)])

arr[arr<10]=11
print(arr)
```
