Dot product

```python
import numpy as np

a=np.array([1,2,3])
b=np.array([4,5,6])
print(np.dot(a,b))
```

matrix multiplication

```python
c=np.array([
    [1,2],
    [3,4]
])

d=np.array([
    [2,3],
    [4,5]
])
result=np.matmul(c,d)
print(result)
```

transpose

```python
arr=np.array([[1,2,3],[5,6,7]])
print(arr.T)
```

determinant

```python
print(np.linalg.det(d))
```

Inverse

```python
print(np.linalg.inv(d))
```
