list vs array

```python
import numpy as np

lst=[1,2,3,4]
print(lst)

#1D array
arr1=np.array([1,2,3,4,5])
print(arr1)

#2D array
arr2=np.array([
    [2,3,4,5],
    [3,4,5,6]
])
print(arr2)
```

array properties

```python
print(arr2.shape)
print(arr2.size)
print(arr2.ndim)
```

special array

```python
print(np.zeros((3,4)))
print(np.ones((2,3)))
print(np.arange(1,10))
```
